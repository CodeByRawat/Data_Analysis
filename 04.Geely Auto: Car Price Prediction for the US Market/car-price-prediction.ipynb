```python
## this code will give all the numerical columns
def numerical():
    k=[] ## creating a empty list
    for i in cardf: # getting all the elements from the list and checking if they are int or float in the next line
        if cardf[i].dtype == int or cardf[i].dtype == float:
         k=k+[f'{i}'] #adding all the float and int elements in the empty list
    return(k)
```

```python
numerical_features=numerical()
```

```python
numerical_features.remove('car_ID')
numerical_features.remove('symboling')
```

```python
numerical_features
```

```python
## distribution of numerical features
plt.figure(figsize=(10,20))
for i in numerical_features:
    plt.subplot(3,5,numerical_features.index(i)+1)
    sns.histplot(data=cardf[i],bins=20,kde=True)
    plt.title(i)
```

```python
## this code will give all the catgorical columns
def categorical():
    k=[] ## creating a empty list
    for i in cardf: # getting all the elements from the list and checking if they are int or float in the next line
        if cardf[i].dtype != int and cardf[i].dtype != float:
         k=k+[f'{i}'] #adding all the float and int elements in the empty list
    return(k)
```

```python
#getting the overview of the data
cardf.head() 
```

```python
#removing car_ID and price 
new_data=cardf.drop(['CarName','car_ID'],axis=1)
```

```python
new_data.head()
```

```python
new_data=pd.get_dummies(new_data,drop_first=True)
```

```python
new_data.head()
```

```python
# storing the column names of the data 
column_list=new_data.columns
```

```python
print(column_list)
```

```python
#removing the output ie 'price' from the input variable
features=list(set(column_list)-set(['price'])) 
```

```python
print(features)
```

```python
y=cardf['price'].values # storing the output variable in y
```

```python
x=new_data[features].values # storing the input variables in x
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
LRprediction=model.predict(test_x)
```

```python
print(f'mean squared error : {LRmse}')
print(f'r2 square value : {LR_r2_square}')
```

```python
#checking for best k value
for i in range(1, 20):
    # Initialize the KNeighbors Regressor with different k values
    knn = KNeighborsRegressor(n_neighbors=i)
    
    # Fit the model on training data
    knn.fit(train_x, train_y)
    
    # Predict the target variable for the test set
    predict_knn_test = knn.predict(test_x)
    
    # Calculate Mean Squared Error (MSE) for predictions
    mse_knn_test = mean_squared_error(test_y, predict_knn_test)
    
    # Print the k value and corresponding MSE
    print(i, ' ', round(mse_knn_test,2))
```

## From the above output the lowest mean squared error value is when k=1

```python

knn_model = KNeighborsRegressor(n_neighbors=1)
knn_model.fit(train_x, train_y)
```

```python
knn_predict=knn_model.predict(test_x)
```

```python
knn_mse=mean_squared_error(test_y,knn_predict)
knn_rsquare=r2_score(test_y,knn_predict)
```

```python
print(knn_mse)
print(knn_rsquare)
```
