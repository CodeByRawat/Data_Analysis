<strong style="color: #E74C3C;">2.Size üì¶:</strong>

```python
# checking for  unique values in the size column
gps_df['Size'].unique()
```


<p><strong style="color: #0000FF;">1.We see that some sizes are in MB and some are in Kbüßê</strong> </p>
<p><strong style="color: #0000FF;">2.Also there is a entry = 'Varies with device'üßê</strong> </p>
<strong style="color: #0000FF;">Therefore we will convert all of them to MB</strong> 


```python
# Function to convert size to MB
def convert_size(value):
    if isinstance(value, str):  # Ensure it's a string before processing
        if 'M' in value:
            return float(value.replace('M', ''))  # Convert '19M' ‚Üí 19.0
        elif 'k' in value:
            return float(value.replace('k', '')) / 1000  # Convert '201k' ‚Üí 0.201
        elif value == 'Varies with device':
            return np.nan  # Handle 'Varies with device'
    return np.nan  # Handle unexpected cases

# Apply conversion to the column
gps_df['Size'] = gps_df['Size'].apply(convert_size)

```

```python
# dropping missing values from type,current ver,android ver
```

```python
gps_df.dropna(subset=['Type','Current Ver','Android Ver'],inplace=True)
```

```python
gps_df.isnull().sum()
```

```python
## Creating the categories for the no of installs
```

```python
# gps_df["Rating"] = pd.to_numeric(gps_df["Rating"], errors="coerce")
```

```python
def categorical_install(installs):
    if installs<0 :
        return 'no'
    elif installs<10:
        return 'Very Low'
    elif installs<100:
        return 'Low'
    elif installs<1000:
        return 'Moderate'
    elif installs<10000:
        return 'More Than Moderate'
    elif installs<100000:
        return 'High'
    elif installs<1000000:
        return 'Very High'
    else:
        return 'Top notch'
```

```python
gps_df['Install_category']=gps_df['Installs'].apply(categorical_install)
```

```python
gps_df['Install_category'].loc[gps_df['Rating'].isnull()].value_counts()
```

```python
# filing the missing values in the ratings columns based on the install category

gps_df['Rating']=gps_df.groupby('Install_category')['Rating'].transform(lambda x:x.fillna(x.mean()))
```

```python
gps_df['Rating'].unique()
```

# üèÜ **Correlation Overview** üì±

```python
x=gps_df['Rating']
y=gps_df['Size']
z=gps_df['Installs']
p=gps_df['Reviews']
q=gps_df['Type']
r=gps_df['Price']

z_log=np.log10(z)
p_log=np.log(p)
```

```python
pair_plot_data=pd.DataFrame({'Rating': x,
    'Size': y,
    'Installs': z_log,
    'Reviews': p_log,
    'Type': q,
    'Price': r})
```

```python
sns.pairplot(pair_plot_data, hue='Type',palette='Set2')
```
